```python
df = pd.read_csv('nyc-bicycle-counts.csv')
```

```python
df['bridgeName'] = df[['Brooklyn Bridge',	'Manhattan Bridge',	'Williamsburg Bridge',	'Queensboro Bridge']].idxmax(axis=1)
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 


# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(df[['High Temp (째F)','Low Temp (째F)',	'Precipitation']], df['bridgeName'], test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
fig = pyplot.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=X_train.columns,  
                   class_names='bridgeName',
                   filled=True)
```

```python
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
df['bridgeName'] = df[['Brooklyn Bridge',	'Manhattan Bridge',	'Queensboro Bridge']].idxmax(axis=1)
df.head()
```

```python
df.tail()
```

```python
df['Brooklyn Bridge Percent Change'] = (df['Brooklyn Bridge']-df['Brooklyn Bridge'].mean())/df['Brooklyn Bridge'].std()
df['Manhattan Bridge Percent Change'] = (df['Manhattan Bridge']-df['Manhattan Bridge'].mean())/df['Manhattan Bridge'].std()
df['Williamsburg Bridge Percent Change'] = (df['Williamsburg Bridge']-df['Williamsburg Bridge'].mean())/df['Williamsburg Bridge'].std()
df['Queensboro Bridge Percent Change'] = (df['Queensboro Bridge']-df['Queensboro Bridge'].mean())/df['Queensboro Bridge'].std()

df['zScore'] = df[['Brooklyn Bridge Percent Change','Manhattan Bridge Percent Change','Williamsburg Bridge Percent Change', 'Queensboro Bridge Percent Change']].max(axis=1)


conditions = [
    (df['zScore'] == (df['Brooklyn Bridge']-df['Brooklyn Bridge'].mean())/df['Brooklyn Bridge'].std()),
    (df['zScore'] == (df['Manhattan Bridge']-df['Manhattan Bridge'].mean())/df['Manhattan Bridge'].std()),
    (df['zScore'] == (df['Williamsburg Bridge']-df['Williamsburg Bridge'].mean())/df['Williamsburg Bridge'].std()),
    (df['zScore'] == (df['Queensboro Bridge']-df['Queensboro Bridge'].mean())/df['Queensboro Bridge'].std())
    ]

# create a list of the values we want to assign for each condition
values = ['Brooklyn Bridge', 'Manhattan Bridge', 'Williamsburg Bridge', 'Queensboro Bridge']

df['Popular_bridge'] = np.select(conditions, values, default= df['zScore'])
```

```python
df.head()
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(df[['High Temp (째F)','Low Temp (째F)',	'Precipitation']], df['Popular_bridge'], test_size=0.3, random_state=1) # 70% training and 30% test

# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn import tree
fig = pyplot.figure(figsize=(25,20))
_ = tree.plot_tree(clf, 
                   feature_names=X_train.columns,  
                   class_names='bridgeName',
                   filled=True)
```
